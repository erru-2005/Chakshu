```python
import json

# Load the JSON data
with open('students.json', 'r') as f:
    data = json.load(f)

# Clean the contact numbers
for student in data:
    for field in ['studentContact', 'parentNo']:
        #print(student[field])
        #break
        if field in student and ".0" in student[field]:
            #print("Here")
            # Convert to int and then back to string to remove ".0"
            student[field] = str(student[field]).replace(".0","")

# Save the cleaned data
with open('students_cleaned.json', 'w') as f:
    json.dump(data, f, indent=4)

print("Cleaned data saved to 'students_cleaned.json'")
```
